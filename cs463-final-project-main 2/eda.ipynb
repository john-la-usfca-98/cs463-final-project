```python
import os, sys
import pandas as pd

dir = os.listdir("data/color")
array = []
df = pd.DataFrame()

photos = []
condition = []
disease = []
healthy = 0
diseased = ""

for file in dir:
    if file != ".DS_Store":
        dir_photos = os.listdir("data/color/" + file)
        if file.__contains__("healthy"):
            healthy = 1
        else:
            healthy = 0

        diseased = (file.split("___"))[1]
        
        for photo in dir_photos:
           condition.append(healthy)
           disease.append(diseased)
           photos.append(photo)

df['Photo'] = photos
df['Disease'] = disease
df['Condition'] = condition

df

```
